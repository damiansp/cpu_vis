```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageDraw, ImageFont, ImageOps
```

```python
IMG = '../../../img'
FIGSIZE = 3, 3
```

```python
bab = np.array(Image.open(f'{IMG}/baboon.png'))
plt.figure(figsize=FIGSIZE)
plt.imshow(bab);
```

```python
img = Image.open(f'{IMG}/cat.png')
plt.figure(figsize=FIGSIZE)
plt.imshow(img);
```

```python
a = np.array(img)
width, height, C = a.shape
print('width, height, C:', width, height, C)
```

```python
a_flip = np.zeros((width, height, C), dtype=np.uint8)
for i,row in enumerate(a):
    a_flip[width - 1 - i, :, :] = row
```

```python
im_flip = ImageOps.flip(img)
# im_flip = image.transpose(1)
plt.figure(figsize=FIGSIZE)
plt.imshow(im_flip);
```

```python
im_mirror = ImageOps.mirror(img)
plt.figure(figsize=FIGSIZE)
plt.imshow(im_mirror);
```

```python
flip = {
    'FLIP_LEFT_RIGHT': Image.FLIP_LEFT_RIGHT,
    'FLIP_TOP_BOTTOM': Image.FLIP_TOP_BOTTOM,
    'ROTATE_90': Image.ROTATE_90,
    'ROTATE_180': Image.ROTATE_180,
    'ROTATE_270': Image.ROTATE_270,
    'TRANSPOSE': Image.TRANSPOSE, 
    'TRANSVERSE': Image.TRANSVERSE}
```

```python
flip['TRANSVERSE']
```

```python
plt.figure(figsize=(10, 10))
for k, v in flip.items():
    plt.subplot(3, 3, v + 1)
    plt.imshow(img.transpose(v))
    plt.title(k)
```

```python
TOP = 150
BOTTOM = 400
crop = a[TOP:BOTTOM, :, :]
plt.figure(figsize=FIGSIZE)
plt.imshow(crop);
```

```python
LEFT = 150
RIGHT = 400
crop = a[TOP:BOTTOM, LEFT:RIGHT, :]
plt.figure(figsize=FIGSIZE)
plt.imshow(crop);
```

```python
plt.figure(figsize=FIGSIZE)
plt.imshow(img.crop((LEFT, TOP, RIGHT, BOTTOM)));
```

```python
asq = np.copy(a)
asq[TOP:BOTTOM, LEFT:RIGHT, 0:2] = 64

plt.imshow(asq);
```

```python
img_draw = img.copy()
img_fn = ImageDraw.Draw(im=img_draw)
shape = [LEFT, TOP, RIGHT, BOTTOM]
img_fn.rectangle(xy=shape, fill='red')
```

```python
plt.figure(figsize=FIGSIZE)
plt.imshow(img_draw);
```

```python
img_fn.text(xy=(TOP, LEFT), text='le box', fill=(0, 0, 0))
plt.figure(figsize=FIGSIZE)
plt.imshow(img_draw);
```

```python
lena = Image.open(f'{IMG}/lenna.png')
alena = np.array(lena)
alena[TOP:BOTTOM, LEFT:RIGHT, :]  = a[TOP:BOTTOM, LEFT:RIGHT, :]
plt.imshow(alena);
```
