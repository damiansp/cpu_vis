```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
IMG = '../../../img'
FS = 5, 5
```

```python
bab = cv2.imread(f'{IMG}/baboon.png')
plt.figure(figsize=FS)
plt.imshow(cv2.cvtColor(bab, cv2.COLOR_BGR2RGB));
```

```python
cat = cv2.imread(f'{IMG}/cat.png')
plt.figure(figsize=FS)
plt.imshow(cv2.cvtColor(cat, cv2.COLOR_BGR2RGB));
```

```python
h, w, c = cat.shape
h, w, c
```

```python
flip = np.zeros((h, w, c), dtype=np.uint8)
for i, row in enumerate(cat):
    flip[w - 1 - i, :, :] = row
```

```python
plt.figure(figsize=FS)
plt.imshow(cv2.cvtColor(flip, cv2.COLOR_BGR2RGB));
```

```python
plt.figure(figsize=FS)
for i, flipcode in enumerate((0, 1, -1)): # up-down, l-r, both
    flip = cv2.flip(cat, flipcode)
    plt.subplot(2, 2, i + 1)
    plt.imshow(cv2.cvtColor(flip, cv2.COLOR_BGR2RGB))
    plt.title(flipcode)
plt.tight_layout()
```

```python
plt.figure(figsize=FS)
rot = cv2.rotate(cat, 0)
plt.imshow(cv2.cvtColor(rot, cv2.COLOR_BGR2RGB));
```

```python
rotd = {
    'r90': cv2.ROTATE_90_CLOCKWISE,
    'rn90': cv2.ROTATE_90_COUNTERCLOCKWISE,
    'r180': cv2.ROTATE_180}
plt.figure(figsize=FS)
for i, (k, v) in enumerate(rotd.items()):
    plt.subplot(2, 2, i + 1)
    plt.imshow(cv2.cvtColor(cv2.rotate(cat, v), cv2.COLOR_BGR2RGB))
    plt.title(f'{k}: {v}')
```

```python
TOP = 150
BOTTOM = 400
L = 150
R = 400

crop = cat[TOP:BOTTOM, L:R, :]
plt.figure(figsize=FS)
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB));
```

```python
sq = np.copy(cat)
sq[TOP:BOTTOM, L:R, :] = sq[TOP:BOTTOM, L:R, ::-1]
plt.figure(figsize=FS)
plt.imshow(cv2.cvtColor(sq, cv2.COLOR_BGR2RGB));
```

```python
start, end = (L, TOP), (R, BOTTOM)
sqc = np.copy(sq)

cv2.rectangle(sqc, pt1=start, pt2=end, color=(255, 255, 0), thickness=3)
plt.figure(figsize=FS)
plt.imshow(cv2.cvtColor(sqc, cv2.COLOR_BGR2RGB));
```

```python
im = cv2.putText(
    img=sqc,
    text='Meow!',
    org=(10, 500),
    color=(0, 255, 255),
    fontFace=4,
    fontScale=5,
    thickness=2)
plt.figure(figsize=FS)
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB));
```
